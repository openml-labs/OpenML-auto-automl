```python
combined_csv = pd.DataFrame()
```

```python
for direc in os.listdir(results_path):
    results_csv = results_path/direc/"results.csv"
    if results_csv.exists():
        df = pd.read_csv(results_csv)
        df["dataset"] = direc
        combined_csv = pd.concat([combined_csv, df])
```
